```python
import yfinance as yf
import pandas as pd
import requests
```

```python
apple = yf.Ticker("AAPL")
```

```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/data/apple.json"
```

```python
data = requests.get(url)
print ("Status code: ", data.status_code)

with open ("apple.json", "wb") as json_file:
   json_file.write(data.content)
```

```python
#getting the apple stock info
import json
with open ("apple.json") as json_file:
    apple_info = json.load(json_file)

apple_info
```

```python
apple_trends = apple.history(period="max")
apple_trends.reset_index(inplace=True)
apple_trends.head()
```

```python
#plotting open size against date on axis x,y..
apple_trends.plot(x="Date", y="Open")
```

```python
#plotting apple stock dividend overtime
apple_trends.Dividends.plot()
```
